# Problem 8: Monte Carlo Approximation (4 points)


## Problem Statement: 

Write a Python program that approximates the value of π using Monte Carlo approximation. Your program should generate a sequence of random points and use these points to estimate the value of π. The accuracy of the approximation should improve as the number of points increases. Here are some hints:
- Your program should generate random points with x and y coordinates ranging between -1 and 1. This will simulate points within a 2 × 2 square that circumscribes a unit circle centered at the origin (0, 0)
- For each generated point, determine whether it falls inside the unit circle.
A point p = (x, y) is inside the circle if x^2 + y^2 <= 1
- Use the ratio of the number of points that fall inside the circle to the total number of generated points to approximate π. The formula is given by π ≈ 4 × (number of points inside/total number of points).

Plot the approximation error of your Monte Carlo estimator as a function of
the total number of points N. You can use np.pi as the ground truth.


```python
# import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# define helper function(s)

def estimate_pi(N):
    # generate random points
    array_shape = (2, N)
    point_cords = np.random.uniform(-1, 1, size=array_shape)

    # count how many points are in the circle
    in_circle = np.sum(point_cords[0]**2 + point_cords[1]**2 <= 1)

    return (in_circle / N) * 4
```

```python
# generate loss values
n_range = range(1, 10000, 2)
loss_arr = []
for N in n_range:
    loss = np.abs(estimate_pi(N) - np.pi)
    loss_arr.append(loss)
```

```python
# plot loss
plt.xlabel('N')
plt.ylabel('loss')
plt.title('Value Over Array')
plt.plot(n_range, loss_arr, label='Loss over N')
```
